```python
# Import Libraries for text preprocessing
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df = pd.read_csv('web_scraping_data.csv', index_col=None)
```

```python
df.head()
```

```python
# Drop duplicates and row with missing values immediately
df = df.drop_duplicates(subset=['Text'])
df = df.dropna()
```

```python
# Check how many rows left
len(df)
```

```python
# Check how many data we have for each label
df['Label'].value_counts()
```

Seems there's an imbalance amount of data among the labels

```python
# Functions for preprocessing the text
def cleansing(df):
  df_clean=df.str.lower() # lowercase
  df_clean=[re.sub(r"\d+","",i) for i in df_clean] # numbers
  df_clean=[re.sub(r'[^\w]', ' ', i) for i in df_clean] # punctuations
  df_clean=[re.sub(r'\s+',' ',i) for i in df_clean] # spaces
  return df_clean

def tokenize(text):
  tokens = [word_tokenize(i) for i in text]
  return tokens

def remove_stopwords(tokens):
  stop_words = set(stopwords.words('indonesian'))
  filtered_tokens = [[word for word in sentence if not word in stop_words] for sentence in tokens]
  return filtered_tokens

def stemming(tokens):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  stemmed = [[token + " : " + stemmer.stem(token) for token in sentence] for sentence in tokens]
  return stemmed
```

```python
# Clean the text
clean_text = cleansing(df['Text'])
df['CleanText'] = clean_text
```

```python
# Rearrange the columns for better viewing
df = df[['Text', 'CleanText', 'Media', 'Label']]
```

```python
df.head()
```

```python
# Split data to training and testing data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(clean_text, df['Label'], test_size=0.2, random_state=42, stratify=df['Label'])
```

```python
# Tokenization, removing stopwords, and stemming
tokens = tokenize(x_train)
tokens_without_stopwords = remove_stopwords(tokens)
stemmed_tokens = stemming(tokens_without_stopwords)
```

# <h1>Text Representation</h1>

```python
# Preparing for TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x_train_tfidf = vectorizer.fit_transform(x_train)
```

```python
# Preparing for Skipgram
import gensim
from gensim.models import Word2Vec
import numpy as np

model_skipgram = gensim.models.Word2Vec(stemmed_tokens, min_count=3, vector_size=50, sg=1)

columns = [f"feature {i}" for i in range(1, 51)] # 50 vector size
train_skipgram = pd.DataFrame(columns=columns)

for text in stemmed_tokens:
    vector = np.mean([model_skipgram.wv[word] for word in text if word in model_skipgram.wv], axis=0)
    train_skipgram.loc[len(train_skipgram.index)] = pd.Series(vector, index=columns)
```

# <h1>Machine Learning</h1>

```python
# Preparing for model training
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid_svc = {
    'C': [0.1, 1, 10],
    'gamma': [0.01, 0.1, 1]
}

param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'min_samples_split': [2, 5, 10]
}
```

```python
# SVM with text representation TF-IDF
svm_tfidf_model = GridSearchCV(estimator=SVC(), param_grid=param_grid_svc, cv=5, scoring='accuracy')
svm_tfidf_model.fit(x_train_tfidf, y_train)
```

```python
# SVM with text representation Skipgram
svm_skipgram_model = GridSearchCV(estimator=SVC(), param_grid=param_grid_svc, cv=5, scoring='accuracy')
svm_skipgram_model.fit(train_skipgram, y_train)
```

```python
# Random Forest Classifier with text representation TF-IDF
rf = RandomForestClassifier(random_state=42)
rf_tfidf_model = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5, scoring='accuracy')
rf_tfidf_model.fit(x_train_tfidf, y_train)
```

```python
# Random Forest Classifier with text representation Skipgram
rf_skipgram_model = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5, scoring='accuracy')
rf_skipgram_model.fit(train_skipgram, y_train)
```

```python
# Preparing the testing dataset for model eveluation (TF-IDF)
x_test_tfidf = vectorizer.transform(x_test)
```

```python
# Predict the result (TF-IDF)
test_svm_tfidf = svm_tfidf_model.predict(x_test_tfidf)
test_rf_tfidf = rf_tfidf_model.predict(x_test_tfidf)
```

```python
# Preparing the testing dataset for model evaluation (Skipgram)
test_skipgram = pd.DataFrame(columns=columns)
```

```python
# Preprocess the testing dataset (Skipgram)
x_test = tokenize(x_test)
x_test = remove_stopwords(x_test)
x_test = stemming(x_test)
```

```python
# Calculates the average word vector for each text using Skipgram model and stores the resulting vector in a Dataframe
for text in x_test:
    vector = np.mean([model_skipgram.wv[word] for word in text if word in model_skipgram.wv], axis=0)
    test_skipgram.loc[len(test_skipgram.index)] = pd.Series(vector, index=columns)
```

```python
# Predict the result (Skipgram)
test_svm_skipgram = svm_skipgram_model.predict(test_skipgram)
test_rf_skipgram = rf_skipgram_model.predict(test_skipgram)
```

# <h1>Model Evaluation</h1>
